```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/Churn_Modelling.csv')
X=dataset.iloc[:,3:13]
y=dataset.iloc[:,13]
```

```python
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
```

```python
X=pd.concat([X,geography,gender],axis=1)
```

```python
X=X.drop(['Geography','Gender'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# now let's make the ANN !!
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
classifier=Sequential()
```

```python
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu',input_dim=11))
classifier.add(Dropout(0.3))
```

```python
classifier.add(Dense(units = 12, kernel_initializer = 'he_uniform', activation = 'relu'))
classifier.add(Dropout(0.4))
```

```python
classifier.add(Dense(units = 8, kernel_initializer = 'he_uniform', activation = 'relu'))
classifier.add(Dropout(0.3))
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
#compiling the ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
mode_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
 y_pred=classifier.predict(X_test)
 y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
