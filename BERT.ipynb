```python
import kagglehub

# Download latest version
path = kagglehub.model_download("tensorflow/bert/tensorFlow2/bert-en-uncased-l-10-h-128-a-2")

print("Path to model files:", path)
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text # Import tensorflow_text to register the necessary ops
from keras import layers

preprocess_url='https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3'
encoder_url= 'https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4'
```

```python
bert_preprocess_model= hub.KerasLayer(preprocess_url)
```

```python
text_test=['nice movie indeed','I love python programming']
text_preprocessed= bert_preprocess_model(text_test)
text_preprocessed.keys()
```

```python
text_preprocessed['input_mask']

#first sentence is of 3 words but in mask there are 5 1's -> CLS word1 word2 word3 SEP
```

```python
bert_model=hub.KerasLayer(encoder_url)
```

```python
bert_result=bert_model(text_preprocessed)  # word embeddings generation
```
